### lamba function

#### short hand functions

```python
n = 3
p = 2
```

```python
def test(n,p):
    return n**p
```

```python
test(3,2)
```

```python
test(9,4)
```

```python
a = lambda n,p :n**p
```

```python
a(9,4)
```

```python
add = lambda x,y : x+y
```

```python
add(234,23)
```

```python
c_to_f = lambda c : (9/5)*c+32
```

```python
c_to_f(34)
```

```python
max_two = lambda x,y : x if x > y else y
```

```python
max_two(4,79)
```

### Map, Reduce, Filter

#### one kind of short hand functions

```python
l = [1,2,3,4,5,45]
```

```python
def test1(l):
    l1 = []
    for i in l:
        l1.append(i**2)
    return l1 
```

```python
test1(l)
```

```python
l1 = [1,2,3,4,5]
l2 = [6,7,8,9,10]
```

```python
list(map(lambda x,y : x+y , l1,l2))
```

```python
f = lambda x,y : x*y
```

```python
list(map(f,l1,l2))
```
