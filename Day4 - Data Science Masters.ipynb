```python
# Assignment
# Sum of Even numbers and odd numbers
lst=[1,2,3,4,5,6,7,8]
```

```python
for i in lst, j in lst:
    if i%2==0:
        if j%2!=0:
         print(i+j)
```

```python
# The "outer" loop to pick 'i'
for i in lst:
    # The "inner" loop to pick 'j'
    for j in lst:
        # Now, put your combined check from your second attempt here
        if i % 2 == 0 and j % 2 != 0:
            # If 'i' is even AND 'j' is odd, print their sum
            print(i + j)
```

```python
even_sum=0
odd_sum=0

for i in lst:
    if i % 2 == 0:
        even_sum += i
    else:
        odd_sum += i

print(even_sum)
print(odd_sum)
```
